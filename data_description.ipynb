Vorhersage treffen:
Das Datenfeld bei dem ich eine Vorhersage treffen möchte ist grades_max (den maximalen erreichten Klettergrad der einzelnen Athleten). Ziel diese ML-Models wird sein basierend auf den Merkmalen des Kletterers oder der Kletterin (Alter, Größe, Gewicht, Geschlecht und Kletterjahre) vorherzusagen, was das maximale-Kletterniveau dieser Person ist.

```python
# 2.2 Statistische Informationen
import csv
import statistics

def read_column(filename, column_name):
    values = []
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file, delimiter=';')
        for row in reader: 
            # Text in Kommazahl (float) unmzuwandeln
            try:
                values.append(float(row[column_name]))
            except ValueError:
                pass # Leere und fehlerhafte Felder werden hiermit übersprungen
        return values

# Daten einlesen
ages = read_column('climber_df.csv', 'age')
heights = read_column('climber_df.csv', 'height')
grades = read_column('climber_df.csv', 'grades_max')
weights = read_column('climber_df.csv', 'weight')
ClimbingYears = read_column('climber_df.csv', 'years_cl')

# Statistiken berechnen und Ausgeben
print("Die Statistische Auswertung, Teilauftrag 2.2")

print(f"Alter -> Median: {statistics.median(ages):.1f}, "
      f"Standard Abweichung: {statistics.stdev(ages):.2f}")
print(f"Grösse -> Median: {statistics.median(heights):.1f}, "
      f"Standard Abweichung: {statistics.stdev(heights):.2f}")
print(f"Grades -> Median: {statistics.median(grades):.1f}, "
      f"Standard Abweichung: {statistics.stdev(grades):.2f}")
print(f"Gewicht -> Median: {statistics.median(weights):.1f}, "
      f"Standard Abweichung: {statistics.stdev(weights):.2f}")
print(f"Kletterjahre -> Median: {statistics.median(ClimbingYears):.1f}, "
      f"Standard Abweichung: {statistics.stdev(ClimbingYears):.2f}")
```

```python
# 2.3 Grafiken

import matplotlib.pyplot as plt

# Grafik 1: Histogramm (Altersverteilung)
print("Grafik 1:")
plt.hist(ages, bins=20, edgecolor='black', color='skyblue')
plt.title('Altersverteilung der Kletterer')
plt.xlabel('Alter in Jahren')
plt.ylabel('Anzahl der Personen')
plt.show()  # zeichnet das erste Bild ins Notebook

# Grafik 2: Streudiagramm (Alter vs. Klettergrad)
print("Grafik 2:")
plt.scatter(ages, grades, alpha=0.5, color='coral')
plt.title('Zusammenhang: Alter und Klettergrad')
plt.xlabel('Alter in Jahren')
plt.ylabel('Maximaler Klettergrad')
plt.show()  # Dieser Befehl zeichnet das zweite Bild ins Notebook
```
